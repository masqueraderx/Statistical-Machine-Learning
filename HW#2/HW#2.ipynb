```python
from sklearn.datasets import fetch_openml
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn import neighbors
```

```python
x, y = fetch_openml('mnist_784', return_X_y=True)
X = np.uint8(x)
Y = np.uint8(y)
x=X.tolist()
y=Y.tolist()
datax=[]
datay=[]
for idx,num in enumerate(y):
    if num==8 or num==9:
        datax.append(x[idx])
        datay.append(num)
datax=np.array(datax)
datay=np.array(datay)
digit_pixels1=datax[0].reshape(28,28)
digit_pixels2=datax[1].reshape(28,28)
digit_pixels3=datax[2].reshape(28,28)
digit_pixels4=datax[4].reshape(28,28)
digit_pixels5=datax[80].reshape(28,28)
digit_pixels6=datax[174].reshape(28,28)
digit_pixels7=datax[77].reshape(28,28)
digit_pixels8=datax[241].reshape(28,28)
plt.subplot(241)
plt.imshow(digit_pixels1)
plt.title('(a) Label=9')
plt.axis('off')
plt.subplot(242)
plt.title('(b) Label=8')
plt.imshow(digit_pixels2)
plt.axis('off')
plt.subplot(243)
plt.title('(c) Label=9')
plt.imshow(digit_pixels3)
plt.axis('off')
plt.subplot(244)
plt.title('(d) Label=8')
plt.imshow(digit_pixels4)
plt.axis('off')
plt.subplot(245)
plt.title('(e) Label=8')
plt.imshow(digit_pixels5)
plt.axis('off')
plt.subplot(246)
plt.title('(f) Label=8')
plt.imshow(digit_pixels6)
plt.axis('off')
plt.subplot(247)
plt.title('(g) Label=8')
plt.imshow(digit_pixels7)
plt.axis('off')
plt.subplot(248)
plt.title('(h) Label=8')
plt.imshow(digit_pixels8)
plt.axis('off')
plt.savefig('Q1.jpg')
```

```python
X_train=datax[:3000]
Y_train=datay[:3000]
X_test=datax[3001:7001]
Y_test=datay[3001:7001]
count=np.zeros(2)
for y in Y_train:
    if y==8:
        count[0]+=1
    else:
        count[1]+=1
print('The ratio of each class is: ',count/3000)
NearestNeighbors = 4
KNNClassifier = neighbors.KNeighborsClassifier(n_neighbors=NearestNeighbors)
KNNClassifier.fit(X_train,Y_train)
predict_Y_train=KNNClassifier.predict(X_train)
Eerror=0
for i in range(3000):
    if predict_Y_train[i]!=Y_train[i]:
        Eerror+=1
EmpiricalRisk=Eerror/3000

predict_Y_test=KNNClassifier.predict(X_test)
Terror=0
WrongX=[]
WrongY=[]
for i in range(4000):
    if predict_Y_test[i]!=Y_test[i]:
        Terror+=1
        WrongX.append(X_test[i])
        WrongY.append([Y_test[i],predict_Y_test[i]])
WrongX=np.array(WrongX)
TestRisk=Terror/4000
digit_pixels1=WrongX[20].reshape(28,28)
digit_pixels2=WrongX[35].reshape(28,28)
digit_pixels3=WrongX[28].reshape(28,28)
digit_pixels4=WrongX[10].reshape(28,28)
print(WrongY[20],WrongY[35],WrongY[28],WrongY[10])
plt.subplot(221)
plt.imshow(digit_pixels1)
plt.title('(a) Label=9,Prediction=8')
plt.axis('off')
plt.subplot(222)
plt.title('(b) Label=8,Prediction=9')
plt.imshow(digit_pixels2)
plt.axis('off')
plt.subplot(223)
plt.title('(c) Label=8,Prediction=9')
plt.imshow(digit_pixels3)
plt.axis('off')
plt.subplot(224)
plt.title('(d) Label=8,Prediction=9')
plt.imshow(digit_pixels4)
plt.axis('off')
plt.tight_layout()
plt.savefig('Q2.jpg')
print('The Empirical Risk is: ',EmpiricalRisk)
print('The Test Risk is: ',TestRisk)
```

```python
Erisk=[]
Trisk=[]
for k in range(1,11):
    KNNClassifier = neighbors.KNeighborsClassifier(n_neighbors=k)
    KNNClassifier.fit(X_train,Y_train)
    predict_Y_train=KNNClassifier.predict(X_train)
    Emerror=0
    for i in range(3000):
        if predict_Y_train[i]!=Y_train[i]:
            Emerror+=1
    EmpiricalRisk=Emerror/3000
    predict_Y_test=KNNClassifier.predict(X_test)
    Testerror=0
    for i in range(4000):
        if predict_Y_test[i]!=Y_test[i]:
            Testerror+=1
    TestRisk=Testerror/4000
    Erisk.append(EmpiricalRisk)
    Trisk.append(TestRisk)
fid = plt.figure()
Axes=plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)
Axes.invert_xaxis()
plt.minorticks_on()
Axes.set_facecolor((0,0,0,0.02))
plt.plot(np.arange(10,0,-1),Erisk[::-1],'o-',markersize=1,color='r',label='Emprical Risk')
plt.plot(np.arange(10,0,-1),Trisk[::-1],'s-',markersize=1,color='b',label='Test Risk')
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlabel("$K$")
plt.ylabel("Risk")
plt.legend(loc='best',fontsize='x-small')
plt.savefig('Q3.jpg')
```

```python
from sklearn.manifold import Isomap
embedding = Isomap(n_components=2)

X_transformed_train=embedding.fit_transform(X_train)
X8_train=[]
Y8_train=[]
X9_train=[]
Y9_train=[]
for idx,label in enumerate(Y_train):
    if label == 8:
        X8_train.append(X_transformed_train[idx])
        Y8_train.append(label)
    else:
        X9_train.append(X_transformed_train[idx])
        Y9_train.append(label)
        
X8_train=np.array(X8_train)
Y8_train=np.array(Y8_train)
X9_train=np.array(X9_train)
Y9_train=np.array(Y9_train)
Axes=plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)
plt.minorticks_on()
Axes.set_facecolor((0,0,0,0.02))
plt.plot(X8_train[:,0],X8_train[:,1],'.',markersize=1,color='r',label='Y=8')
plt.plot(X9_train[:,0],X9_train[:,1],',',markersize=1,color='b',label='Y=9')
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlabel("First Dimension")
plt.ylabel("Second Dimension")
plt.legend(loc='best',fontsize='x-small',framealpha=0.4)
# plt.show()
plt.savefig('Q4.jpg')
```
