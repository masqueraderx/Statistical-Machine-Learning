```python
from sklearn.svm import SVR
import numpy as np
from matplotlib import pyplot as plt
```

# $\epsilon$=100

```python
# generate the dataset
np.random.seed(2020) ;
X = np.arange(100)
Y = 2*X + 50*np.random.randn(X.size)
eps = 100
# set the model
Model = SVR(C=1e7, epsilon=eps, kernel="linear")
Model.fit(X.reshape(-1,1),Y)
# prediction
y_predict = Model.predict(X.reshape(-1,1))
```

```python
fid = plt.figure(figsize=(10,5))

Axes = plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)
plt.minorticks_on()
Axes.set_facecolor((0,0,0,0.02))
# draw the classifier y
plt.plot(X, y_predict, 'k-', linewidth=2, color = 'blue', label='$y_{pred}$')
# draw the y + epsilon
plt.plot(X, y_predict+eps, 'k--', linewidth=2, color = 'blue', label='$y_{pred}+\epsilon$')
# draw the y - epsilon
plt.plot(X, y_predict-eps, 'k--', linewidth=2, color = 'blue', label='$y_{pred}-\epsilon$')
# draw the support vector
plt.scatter(X[Model.support_], Y[Model.support_], marker = 's', color = 'blue', label = "vector support")
# draw the original dataset
x = np.delete(X, [Model.support_])
y = np.delete(Y, [Model.support_])
plt.plot(x, y, 'o', color='r', label='data')
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlabel("$x$ coordinate")
plt.ylabel("$y$ coordinate")
plt.legend(loc='lower right',fontsize='x-small')
plt.savefig('/Users/gexueren/Desktop/6254/Code/Latex/HW#7/Q26.jpg')
```
