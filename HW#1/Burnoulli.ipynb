```python
from scipy.stats import binom
from matplotlib import pyplot as plt
import numpy as np
import math

n=10000
p=0.4
alpha=0.05
res=[]
output=[]
for j in range(1,n+1):
    epsilon=math.sqrt(math.log(2/alpha)/(2*j))
    interval=2*epsilon
    res.append(interval)
    if interval<=alpha:
        output.append(j)
print(output[0])
plt.plot(np.arange(1,n+1),res,color='b')
plt.plot(np.arange(1,n+1),np.ones((n,1))*(alpha),color='r')
plt.grid(True,linewidth=0.5)
plt.xlabel('$N$')
plt.ylabel('Length of Interval')
plt.title('Length of Interval versus N')
```

```python
from scipy.stats import binom
from matplotlib import pyplot as plt
import numpy as np
import math

n=10000
p=0.4
alpha=0.05
size=10000
# k=np.arange(0,size)
res=[]
for j in range(1,n+1):
    count=0
    p_n=np.random.binomial(j,p,size)/j
#     p_n = binom.cdf(k, j, p)/j
    epsilon=math.sqrt(math.log(2/alpha)/(2*j))
    left_bound=p_n-epsilon
    right_bound=p_n+epsilon
    for i in range(size):
        if left_bound[i]<p<right_bound[i]:
            count+=1
    res.append(count/size)
plt.plot(np.arange(1,n+1),res,markersize=0.5,color='b')
plt.plot(np.arange(1,n+1),np.ones((n,1))*(1-alpha),markersize=0.5,color='r')
plt.grid(True,linewidth=0.5)
plt.xlabel('$N$')
plt.ylabel('$Converge$')
plt.title('Converge Versus N')
```
