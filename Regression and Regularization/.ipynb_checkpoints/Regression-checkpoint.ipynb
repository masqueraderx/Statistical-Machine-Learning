```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression, Ridge
```

# Overfitting in the presence of noise

```python
np.random.seed(20201027)
x = np.random.uniform(-1.0, 1.0, 5)
y = x**2 + np.random.normal(0.0, 0.2, 5)

x_range = np.linspace(-1.0, 1.0, 100)
y_model = (x_range)**2
plt.plot(x_range, y_model, 'b-', label='True Model')
plt.scatter(x, y, c='red', label='Noisy Data')
plt.xlim([-1.0, 1.0])
plt.ylim([-0.01, 1.0])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('True Model vs Noisy Data')
plt.legend()
```

# When there are too many degrees of freedom in model

```python
degree = 6
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(x.reshape(-1,1), y)
y_pred = model.predict(x_range.reshape(-1,1))
plt.plot(x_range, y_model, 'b-', label='True Model')
plt.scatter(x, y, c='red', label='Noisy Data')
plt.plot(x_range, y_pred, 'g-', linewidth=2, label="Degree %d" % degree)
plt.xlim([-1.0, 1.0])
plt.ylim([-0.01, 1.0])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
```

# When hypothesis set contains simpler functions than target functions but few samples

```python
x_range = np.linspace(-1.0, 1.0, 100)
y_model = (x_range)**2
sdev = 0.1
degree = 1
Npoints = 5
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

plt.figure()

for i in np.arange(20):
    x = np.random.uniform(-1.0, 1.0, Npoints)
    y = (x)**2 + np.random.uniform(0.0, sdev, Npoints)
    model.fit(x.reshape(-1,1),y)
    y_pred = model.predict(x_range.reshape(-1,1))
    plt.plot(x_range, y_pred, 'g-', linewidth=2)

plt.plot(x_range, y_model, 'b--', label='True Model')
plt.xlim([-1.0, 1.0])
plt.ylim([-0.2, 1.0])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
```

# Ridge Regression (Regularization)

```python
x_range = np.linspace(-1.0, 1.0, 100)
y_model = (x_range)**2
sdev = 0.1
degree = 1
Npoints = 5
model = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=1.0))

plt.figure()

for i in np.arange(20):
    x = np.random.uniform(-1.0, 1.0, Npoints)
    y = (x)**2 + np.random.uniform(0.0, sdev, Npoints)
    model.fit(x.reshape(-1,1),y)
    y_pred = model.predict(x_range.reshape(-1,1))
    plt.plot(x_range, y_pred, 'g-', linewidth=2)

plt.plot(x_range, y_model, 'b--', label='True Model')
plt.xlim([-1.0, 1.0])
plt.ylim([-0.2, 1.0])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
```
