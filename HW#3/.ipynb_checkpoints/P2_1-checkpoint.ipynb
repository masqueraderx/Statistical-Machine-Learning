```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn.naive_bayes import GaussianNB
from matplotlib.colors import LinearSegmentedColormap
```

```python
BlueDataSet = np.random.multivariate_normal([1,1],[[1,0],[0,1]],50)   #label 0
RedDataSet = np.random.multivariate_normal([-1,-1],[[1,0],[0,1]],50)  #label 1
```

```python
TrainingData = np.r_[BlueDataSet, RedDataSet]
TrainingLabel = np.r_[np.zeros((50,1)), np.ones((50,1))]
gnb = GaussianNB()
gnb.fit(TrainingData, TrainingLabel)
```

```python
fid = plt.figure()
Axes = plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)
plt.minorticks_on()
# Axes.xaxis.set_ticks_position(position='both')
# Axes.yaxis.set_ticks_position(position='both')
Axes.set_facecolor((0,0,0,0.02))
plt.plot(BlueDataSet[:,0],BlueDataSet[:,1],'o',color='b',label='Blue class')
plt.plot(RedDataSet[:,0],RedDataSet[:,1],'s',color='r',label='Red class')
plt.plot(-1,-0.5,'kd',label='Class mean')
plt.plot(1,0.5,'kd')
BlueClassSdev = matplotlib.patches.Circle((-1,-0.5),radius=2,color='blue',alpha=0.05,label='2$\sigma$ for Blue class')
Axes.add_patch(BlueClassSdev)
RedClassSdev = matplotlib.patches.Circle((1,0.5),radius=2,color='red',alpha=0.05,label='2$\sigma$ for Red class')
Axes.add_patch(RedClassSdev)
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
# plt.xlim([-5,5])
# plt.ylim([-4,4])
plt.xlabel("$x$ coordinate")
plt.ylabel("$y$ coordinate")
plt.legend(loc='lower right',fontsize='x-small')
```

```python
x_grid = np.linspace(-4.1,3,100)
y_grid = np.linspace(-3.1,4,100)
x_mesh, y_mesh = np.meshgrid(x_grid,y_grid)
PredictedLabel = gnb.predict(np.c_[x_mesh.ravel(),y_mesh.ravel()])
Xcoord = x_mesh.ravel()
Ycoord = y_mesh.ravel()
BlueRegionX = Xcoord[PredictedLabel==0]
BlueRegionY = Ycoord[PredictedLabel==0]
RedRegionX = Xcoord[PredictedLabel==1]
RedRegionY = Ycoord[PredictedLabel==1]
# RedRegion = Xcoord[PredictedLabel==-1]
plt.plot(BlueRegionX,BlueRegionY.ravel(),'bo')
plt.plot(RedRegionX,RedRegionY.ravel(),'rs')
```

```python
my_colors = [(0,0,1),(1,0,0)]#RGB
my_cm = LinearSegmentedColormap.from_list('my_cm', my_colors, N=2)
plt.contourf(x_mesh,y_mesh,PredictedLabel.reshape(100,100),levels=[-1,0,1], cmap = my_cm, alpha=0.2)
plt.plot(BlueDataSet[:,0],BlueDataSet[:,1],'o',color='b',label='Blue class')
plt.plot(RedDataSet[:,0],RedDataSet[:,1],'s',color='r',label='Red class')
plt.contour(x_mesh,y_mesh,PredictedLabel.reshape(100,100),linewidth=2,colors='k')
```
