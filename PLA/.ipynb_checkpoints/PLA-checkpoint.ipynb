```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn import linear_model, neighbors, metrics
from matplotlib import patches, cm, animation
from matplotlib.colors import LinearSegmentedColormap
from tqdm import tqdm, tnrange, tqdm_notebook
```

```python
my_colors=[(1, 0, 0), (0, 0, 1)]
my_cm = LinearSegmentedColormap.from_list('my_cm', my_colors, N=2)
```

```python
np.random.seed(2020)
Sigma =[[10,5],[5,5]]
N = 50 #number of data points
blue_data = np.random.multivariate_normal([-4,3], Sigma, N) 
blue_labels = 1 * np.ones((N,1)) #coding blue label as one
red_data = np.random.multivariate_normal([2,-1], Sigma, N) 
red_labels = 0 * np.ones((N,1)) #coding red label as zero
data = np.concatenate((blue_data, red_data), axis=0)
label = np.concatenate((blue_labels, red_labels), axis=0).ravel()
```

```python
#Need to create a mesh to evaluate classifier
x_grid = np.linspace(1.5*np.min(data[:,0]), 1.5*np.max(data[:,0]), 300)
y_grid = np.linspace(1.5*np.min(data[:,1]), 1.5*np.max(data[:,1]), 300)
x_mesh, y_mesh = np.meshgrid(x_grid, y_grid)
```

```python
%matplotlib inline
fid =plt.figure()

Axes = plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)    #direction set the ticks direction
plt.minorticks_on()     #to set minor ticks on
Axes.set_facecolor((0,0,0,0.02))          #facecolor：背景颜色  edgecolor：边框颜色
plt.plot(blue_data[:,0],blue_data[:,1],'o',color='b',label='Blue class')
plt.plot(red_data[:,0],red_data[:,1],'s',color='r',label='Red class')
plt.plot(-4,3,'kd',label='Class mean')
plt.plot(2,-1,'kd')
EigenValues, EigenVectors = np.linalg.eig(Sigma)
OrientationAngle = np.arctan(EigenVectors[1,1]/EigenVectors[0,1])/np.pi*180
BlueClassSdev = matplotlib.patches.Ellipse((-4,3),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='blue',alpha=0.1,label='2$\sigma$ for Blue class')
Axes.add_patch(BlueClassSdev)
RedClassSdev = matplotlib.patches.Ellipse((2,-1),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='red',alpha=0.1,label='2$\sigma$ for Red class')
Axes.add_patch(RedClassSdev)
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlim([1.1*np.min(data[:,0]),1.1*np.max(data[:,0])])
plt.ylim([1.1*np.min(data[:,1]),1.1*np.max(data[:,1])])
plt.xlabel("$x$ coordinate")
plt.ylabel("$y$ coordinate")
plt.legend(loc='lower right',fontsize='x-small')
plt.title("PLA Classification")
plt.savefig('data.jpg')
```

```python
UpdateOrder = np.random.permutation(2*N)
```

```python
theta = np.r_[1, data[UpdateOrder[0]]]
A=np.array([1,data[2][0],data[2][1]])
B=np.dot(-1,[[1],[data[2][0]],[data[2][1]]])
D=np.dot(-1,[1,data[2][0],data[2][1]])
print(D)
C=theta.dot([[1],[data[2][0]],[data[2][1]]])
print(theta)
```

```python
def PLAprediction(xcoord,ycoord,theta):
    pred = np.zeros(xcoord.shape)
    for i in np.arange(0, len(xcoord)):
        lin_class = theta.dot([[1],[xcoord[i]], [ycoord[i]]])
        if lin_class > 0:
            pred[i] = +1
        else:
            pred[i] = -1
    return (pred)

prediction0 = PLAprediction(x_mesh.ravel(), y_mesh.ravel(),theta)
prediction0 = prediction0.reshape(x_mesh.shape)
```

```python
fid =plt.figure()

Axes = plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)    #direction set the ticks direction
plt.minorticks_on()     #to set minor ticks on
Axes.set_facecolor((0,0,0,0.02))          #facecolor：背景颜色  edgecolor：边框颜色
plt.plot(blue_data[:,0],blue_data[:,1],'o',color='b',label='Blue class')
plt.plot(red_data[:,0],red_data[:,1],'s',color='r',label='Red class')
plt.plot(-4,3,'kd',label='Class mean')
plt.plot(2,-1,'kd')
EigenValues, EigenVectors = np.linalg.eig(Sigma)
OrientationAngle = np.arctan(EigenVectors[1,1]/EigenVectors[0,1])/np.pi*180
BlueClassSdev = matplotlib.patches.Ellipse((-4,3),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='blue',alpha=0.1,label='2$\sigma$ for Blue class')
Axes.add_patch(BlueClassSdev)
RedClassSdev = matplotlib.patches.Ellipse((2,-1),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='red',alpha=0.1,label='2$\sigma$ for Red class')
Axes.add_patch(RedClassSdev)
PLA = plt.contour(x_mesh,y_mesh,prediction0,[-1,0,1.2],linewidth=2,color='k')
plt.contourf(x_mesh,y_mesh,prediction0,level=[-1,0,1],cmp=my_cm,alpha=0.05)
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlim([1.1*np.min(data[:,0]),1.1*np.max(data[:,0])])
plt.ylim([1.1*np.min(data[:,1]),1.1*np.max(data[:,1])])
plt.xlabel("$x$ coordinate")
plt.ylabel("$y$ coordinate")
plt.legend(loc='lower right',fontsize='x-small')
plt.title("PLA Classification")
plt.savefig('before.jpg')
```

```python
%matplotlib inline
for CurrentUpdate in UpdateOrder[1:]:
    CurrentPrediction = theta.dot([[1],[data[CurrentUpdate][0]],[data[CurrentUpdate][1]]])
#     print(np.sign(CurrentPrediction))
    if np.sign(CurrentPrediction) != 2*label[CurrentUpdate]-1:
        theta = theta + (2*label[CurrentUpdate]-1) * np.r_[1, data[CurrentUpdate]]
        prediction = PLAprediction(x_mesh.ravel(), y_mesh.ravel(),theta)
        prediction = prediction.reshape(x_mesh.shape)
        
fid =plt.figure()

Axes = plt.subplot(1,1,1)
Axes.axes.tick_params(which='both',direction='in',top=True, right=True)    #direction set the ticks direction
plt.minorticks_on()     #to set minor ticks on
Axes.set_facecolor((0,0,0,0.02))          #facecolor：背景颜色  edgecolor：边框颜色
plt.plot(blue_data[:,0],blue_data[:,1],'o',color='b',label='Blue class')
plt.plot(red_data[:,0],red_data[:,1],'s',color='r',label='Red class')
plt.plot(-4,3,'kd',label='Class mean')
plt.plot(2,-1,'kd')
EigenValues, EigenVectors = np.linalg.eig(Sigma)
OrientationAngle = np.arctan(EigenVectors[1,1]/EigenVectors[0,1])/np.pi*180
BlueClassSdev = matplotlib.patches.Ellipse((-4,3),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='blue',alpha=0.1,label='2$\sigma$ for Blue class')
Axes.add_patch(BlueClassSdev)
RedClassSdev = matplotlib.patches.Ellipse((2,-1),width=EigenValues[1],height=EigenValues[0],angle=OrientationAngle,color='red',alpha=0.1,label='2$\sigma$ for Red class')
Axes.add_patch(RedClassSdev)
PLA = plt.contour(x_mesh,y_mesh,prediction,[-1,0,1.2],linewidth=2,color='k')
plt.contourf(x_mesh,y_mesh,prediction,level=[-1,0,1],cmp=my_cm,alpha=0.05)
plt.grid(True,which='major',linewidth=0.5)
plt.grid(True,which='minor',linewidth=0.1)
plt.xlim([1.1*np.min(data[:,0]),1.1*np.max(data[:,0])])
plt.ylim([1.1*np.min(data[:,1]),1.1*np.max(data[:,1])])
plt.xlabel("$x$ coordinate")
plt.ylabel("$y$ coordinate")
plt.legend(loc='lower right',fontsize='x-small')
plt.title("PLA Classification")
plt.savefig('after.jpg')
```
